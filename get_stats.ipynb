```python
import pandas as pd
import numpy as np
import os
import re
```

```python
def get_stats(df) :
    word_count = 0
    for i in df['full_text'] :
        word_count+= len(re.findall(r'\w+', i))
    return word_count

def get_datewise(df,date,month,country,outputF) :
    df_temp = df[(df['month']==month) & (df['date']==int(date))]
    temp = df_temp[(df_temp['country']==country) | (df_temp['location']==country)]
    if(country == 'united states') :
        suffix = 'usa'
    elif(country == 'india'):
        suffix = 'ind'
    elif(country=='united kingdom'):
        suffix = 'uk'
    elif(country == 'singapore') :
        suffix = 'sgp'
    tweet_count = 0
    avg_words_per_tweet = 0 
    if(len(temp)>0):
        print("tweets in "+country+ " = " + str(len(temp)))
        tweet_count = len(temp)
        avg_words_per_tweet = (get_stats(temp)/tweet_count)
        output_file =outputF+suffix+".csv"
        temp.to_csv(output_file)
    
    return tweet_count , avg_words_per_tweet
```
