```python
import pandas as pd
import numpy as np
import os
import re
```

```python
def get_stats(df) :
    word_count = 0
    for i in df['full_text'] :
        word_count+= len(re.findall(r'\w+', i))
    return word_count

def get_datewise(df,date,month,country,suffix,outputF) :
    temp = df[(df['month']==month) & (df['date']==int(date))]
    tweet_count = 0
    avg_words_per_tweet = 0 
    if(len(temp)>0):
        print("tweets in "+country+ " = " + str(len(temp)))
        tweet_count = len(temp)
        avg_words_per_tweet = (get_stats(temp)/tweet_count)
        output_file =outputF+suffix+".csv"
        temp.to_csv(output_file)
    
    return tweet_count , avg_words_per_tweet
```

```python
region = 'ven'
# german : de spanish : es french:fr italian :it dutch:nl russian:ru
lang = 'es'
country = 'venezuela'
```

```python
flag = 0
completed_tweets_path = '../new_csvs/'+region+'/'
for _,_,files in os.walk(completed_tweets_path) :
    print(files)
print(len(files))
total_tweets = 0

for file in files :
    if(flag==0) :
        df=pd.read_csv(completed_tweets_path+file, index_col=False)
        flag=1
        print(len(df))
        total_tweets+=len(df)
    else :
        df_temp = pd.read_csv(completed_tweets_path+file , index_col=False, error_bad_lines=False)
        print(len(df_temp))
        df = pd.concat([df,df_temp])
        total_tweets+=len(df_temp)
```

```python
df = df.reset_index(drop = True)
df
```

```python
df.drop_duplicates(subset='tweet_id', keep='first', inplace=True)
```

```python
df = df[df['language']==lang]
df
```

```python
months = ['Jan' ,'Feb' , 'Mar', 'Apr', 'May']
days = {
    'Jan' : 31,
    'Feb' : 29,
    'Mar' : 31,
    'Apr' : 30,
    'May' : 18,
}
```

```python
tweets = []
wpt_stat = []
mth = []
dt = []
tweet_sum = 0

for m in months :
    for i in range(1,days[m]+1) :
        print(m)
        print(i)
        if(i <=9) :
            date = '0'+str(i)
        else :
            date= str(i)
        
        mth.append(m)
        dt.append(i)
        
        outputF = "../datewise_countrywise/"+region+"/"+ m + "_" +date+"_"
        tot_tweet , wpt = get_datewise(df,date,m,country,region,outputF)
        tweet_sum+=tot_tweet
        tweets.append(tot_tweet)
        wpt_stat.append(wpt)
```

```python
print(tweet_sum)
```

```python
stats_arg = {
    'date': dt,
    'month': mth,
    'tweets_arg' :tweets,
    'wpt_arg' : wpt_stat
}
```
